```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
def transform(x,d):
    poly=np.ones((len(x),1))
    for i in range(1,d):
        poly=np.c_[poly,np.power(x,i)]
    return poly
```

```python
def coefficient_of_poly(x,y):
    xnew=transform(x,len(x))
    return np.dot(np.linalg.inv(xnew),y.reshape(-1,1),)
```

```python
a=5
x=np.linspace(1,10,a)
print(x.shape)
y = np.random.rand(a,)
plt.bar(x,y, color='blue', label='Actual data')
x1=np.linspace(1,10,100).reshape(100,1)
coef=coefficient_of_poly(x,y)
print(coef.shape)
print(transform(x1,a).shape)
plt.plot(x1,np.dot(transform(x1,a),coefficient_of_poly(x,y)), color='red', label='Polynom')
plt.xlabel('X')
plt.ylabel('y')
plt.title('interpolation pour polynomes')
plt.legend()
plt.show()


```
